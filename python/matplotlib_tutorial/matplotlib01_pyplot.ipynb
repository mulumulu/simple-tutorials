# filling_between: 二つの折れ線の間に色を塗る

```python
'''
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.fill_between.html
'''
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y1= np.sin(x)
plt.plot(x, y1)
plt.fill_between(x, y1)
plt.show()
```

```python
y2 = np.cos(x)
plt.plot(x, y1)
plt.plot(x, y2)
plt.fill_between(x, y1=y1, y2=y2)
plt.show()
```

```python
n = 10
x = np.linspace(-np.pi, np.pi, n)
y = np.sin(x)
plt.plot(x, y)

## パラメータの設定
where = [True]*(n//2) + [False]*(n//2)
interpolate = True

plt.fill_between(x, y1=y, where=where, interpolate=interpolate)
plt.show()
```

# countourf: 等高線

```python
'''
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.contourf.html
'''
```

```python
n = 100
x = np.linspace(-1, 1, n)
y = np.linspace(-1, 1, n)
xx, yy= np.meshgrid(x,y)
zz = xx**2 + yy**2
plt.contourf(xx, yy, zz, levels=100)
plt.colorbar()
```
